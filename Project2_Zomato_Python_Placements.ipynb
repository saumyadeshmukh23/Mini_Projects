```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('/content/drive/MyDrive/zomato.csv')
```

```python
data.head()
```

```python
data = data.drop(['url','address','phone','menu_item', 'listed_in(type)', 'listed_in(city)'],axis=1)
```

```python
data
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
null_percentage = (data.isna().sum() / len(data)) * 100
print(null_percentage)
```

**Data Cleaning**

```python
feature_na = [i for i in data.columns if data[i].isna().sum() > 0]
feature_na
```

```python
for i in feature_na:
    print(f"{i} has {np.round((data[i].isnull().sum()/len(data[i]) * 100), 4)}% null values")
```

```python
data.rate.unique()
```

```python
data.dropna(subset=["rate"], axis=0, inplace=True)
```

```python
def split(x):
  return x.split('/')[0].strip()
```

```python
data['rate'] = data['rate'].apply(split)
data['rate']
```

```python
data.rate.unique()
```

```python
data['rate'].replace(['NEW','-'], 0, inplace=True)
```

```python
data.info()
```

```python
data['rate'] = data['rate'].astype(float)
```

```python
data.info()
```

```python
# data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(float)
```

# ***Calculate avg rating for each restaurant***

```python
rating = pd.pivot_table(data, values='rate', index=['name'])
rating
```

```python
rating = rating.sort_values(['rate'], ascending=False)
rating[0:20]
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=rating[0:20].rate, y=rating[0:20].index, orient='h')
plt.xticks(rotation=90)
plt.show()
```

# ***Distribution of Ratings***


```python
sns.set_style('whitegrid')
sns.distplot(data['rate'])
plt.show()
```

```python
from scipy.stats import normaltest
DataToTest = data['rate']
stat, p = normaltest(DataToTest)
print('Statistics=%.2f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Normal Distribution')
else:
	print('Not a normal distribution')
```

# Lets analyse the highly rated restaurants and worst restaurants based on number of ratings given

```python
data.head(3)
```

```python
sns.distplot(data['votes'])
plt.show()
```

```python
voting_rating = data.groupby('name')[['votes']].mean().sort_values('votes', ascending=False)
voting_rating
```

```python
voting_rating['name'] = voting_rating.index
```

```python
voting_rating = voting_rating.reset_index(drop=True)
voting_rating.head()
```

```python
voting_rating = pd.merge(voting_rating, data[['rate', 'name']])
voting_rating
```

```python
vote_top = voting_rating[voting_rating['votes'] > 5000]
vote_top.head()
```

```python
vote_top = vote_top.groupby('name')[['rate']].mean().sort_values('rate', ascending=False)
vote_top
```

# Highest Rated based on most number of votes

```python
plt.figure(figsize=(20, 8), dpi=100)
sns.barplot(x = vote_top['rate'], y = vote_top.index, orient='h', palette='deep')
plt.show()
```

```python
bad_count = voting_rating[(voting_rating['rate'] < 3) & (voting_rating['rate'] > 0)]
bad_count
```

```python
bad_restaurants = bad_count[bad_count['votes'] > 500].groupby('name')[['rate']].mean()
bad_restaurants
```

# Worst Restaurants based on highest number of votes given

```python
plt.figure(figsize=(20, 8))
sns.barplot(x=bad_restaurants['rate'], y=bad_restaurants.index, palette='deep')
plt.show()
```
