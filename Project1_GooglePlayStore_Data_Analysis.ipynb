```python
def outlier_analysis():
  q1 = data["Rating"].quantile(0.25)
  q3 = data["Rating"].quantile(0.75)
  IQR = q3 - q1
  df1 = data["Rating"]
  UF = q3 + (1.5 * IQR)
  LF = q1 - (1.5 * IQR)
  outliers = df1[(df1 > UF) | (df1 < LF)]
  return outliers

outlier_analysis()
```
